```python
print("Yahoo".upper())
print()
print("Sign in to Yahoo Mail".upper())
print()
print("Using your Yahoo Account".upper())
print()

Email = input("Enter your Email ID: ")

for attempt in range(3):
    Password = input("Enter your Password: ")

    valid_length = False
    lower_case = False
    upper_case = False
    digits = False
    special_characters = False

    if 8 <= len(Password) <= 16:
        valid_length = True

        for char in Password:
            if char.islower():
                lower_case = True

            if char.isupper():
                upper_case = True

            if char.isdigit():
                digits = True

            if char in "@$_#^&*":
                special_characters = True

    if valid_length and lower_case and upper_case and digits and special_characters:
        print("Valid Password")
        break
    else:
        print("Invalid Password. Try again.")

    if attempt == 2:
        print("You have used all your attempts. Access denied.")
    else:
        print("Too many failed attempts. Generating a new password...")      

```
